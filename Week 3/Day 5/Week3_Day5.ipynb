```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error

df=pd.read_csv('/content/seep_mela_train_clean.csv')
print(df)
```

```python
dfcopy=df.copy()
```

```python
le = LabelEncoder()
dfcopy['Course'] = le.fit_transform(dfcopy['Course'])
```

```python
print(dfcopy['Course'])
```

```python
x = np.column_stack((dfcopy[['Year']],dfcopy['Course'],dfcopy['Sessions_Conducted']))
y = np.array(dfcopy['Participants'])
```

```python
#TRAIN-TEST SPLIT
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.predict([[2082,4,5]])
```

```python
params={'alpha':[0.1,1.0,10.0,100.0]}
```

```python
grid=GridSearchCV(Ridge(),params,cv=5,scoring='neg_mean_squared_error')
grid.fit(X_train,y_train)
```

```python
print('Best Hyperparameters:',grid.best_params_)
```

```python
best_alpha=grid.best_params_['alpha']
final_model=Ridge(alpha=best_alpha)
final_model.fit(X_train,y_train)
```
